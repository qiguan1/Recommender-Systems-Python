```python
import gzip
import collections
from collections import defaultdict
import random
import string
import scipy.optimize
from sklearn import svm
import numpy as np
import pandas as pd 

def readGz(f):
  for l in gzip.open(f):
    yield eval(l)
    
def pearson(i,j):
    ii=i-i.mean()
    jj=j-j.mean()
    return np.sum(ii*jj)/np.sqrt(np.sum(ii**2)*np.sum(jj**2))
  

user_list_train= []
business_list_train = []
rating_list_train = []
user_list_all= []
business_list_all = []
itr=0
for l in readGz("train.json.gz"): 
  user_list_train.append(l['userID'])
  business_list_train.append(l['businessID'])
  rating_list_train.append(l['rating'])
  itr+=1
  if itr == 100000: break

for l in readGz("train.json.gz"): 
  user_list_all.append(l['userID'])
  business_list_all.append(l['businessID'])

ub_pair_train=list(zip(user_list_train,business_list_train))

d_u_train = defaultdict(list) 
for u,b in ub_pair_train:
  d_u_train[u].append(b)

ub_pair=list(zip(user_list_all,business_list_all))

d_u = defaultdict(list) 
for u,b in ub_pair:
  d_u[u].append(b)

ub_positive=ub_pair[int(len(ub_pair)/2):]

ub_negative = []
ub_negative_dict=defaultdict(list)

count = 0
while(count < 100000):
  user_rand = user_list_all[random.randint(0,199999)]
  business_rand = business_list_all[random.randint(0,199999)]
  if business_rand in d_u[user_rand]:
    continue
  else:
    if business_rand not in ub_negative_dict[user_rand]:
      count += 1  
      ub_negative_dict[user_rand].append(business_rand)
      ub_negative.append((user_rand, business_rand))
    
ub_validation = ub_positive + ub_negative

data_list = {'userID': user_list_train, 'businessID': business_list_train, 'rating': rating_list_train}
data_list = collections.OrderedDict(data_list)
ratings = pd.DataFrame.from_dict(data=data_list)
matrix = ratings.pivot_table(index='userID',columns='businessID',values='rating')
print("done")





```

```python
predict = []

for l in ub_validation:
  u,b = l[0],l[1]
  if b not in matrix.columns:
    predict.append(0)
  else:
    p_list=[]
    count = 0
    for i in d_u_train[u]:
      if count > 5: break
      p_list.append(pearson(matrix[b],matrix[i]))
      count += 1
    
    p_list = np.array(p_list)
    cor = p_list.mean()

    if cor > 0.3:
      predict.append(1)
    else:
      predict.append(0)

acc = 0
p1 = predict[:int(len(predict)/2)]
p2 = predict[int(len(predict)/2):]
for p in p1:
  if p == 1: acc +=1

for p in p2:
  if p == 0: acc +=1


print("Accuracy on validation set is: " + str(acc/len(ub_validation)))


```
